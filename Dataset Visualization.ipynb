```python
plt.figure(figsize=(12,12))

plt.subplot(221)
sns.scatterplot(data=iris_df, x='sepallength', y='sepalwidth', hue='target')

plt.subplot(222)
sns.scatterplot(data=iris_df, x='petallength', y='petalwidth', hue='target')

plt.subplot(223)
sns.scatterplot(data=iris_df, x='sepallength', y='petallength', hue='target')

plt.subplot(224)
sns.scatterplot(data=iris_df, x='sepalwidth', y='petalwidth', hue='target')

plt.show()
```

```python
cor = iris_df.drop('target', axis=1).corr(method='pearson')
cor
```

```python
sns.heatmap(cor, annot=True)
```

Good correlation between:
1. petalwidth and petallength -> 0.96
2. petalwidth and sepallength -> 0.82
3. petallength and sepallength -> 0.87
